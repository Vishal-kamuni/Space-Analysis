```python
# Drop irrelevant columns
df = df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
```

```python
# Rename Column
df = df.rename(columns={"Status Rocket": "Rocket_Status", " Rocket": "cost of the rocket launch($M)", "Status Mission": "Mission_Status","Detail":"Rocket name and satellites "})
```

```python
# Convert "Datum" to datetime format
df["Datum"] = pd.to_datetime(df["Datum"], errors='coerce')
```

```python
df.dtypes
```

```python
#Convert "Cost of the rocket launch" column to numeric
df["cost of the rocket launch($M)"] = pd.to_numeric(df["cost of the rocket launch($M)"], errors='coerce')
```

```python
#handle missing values
df["cost of the rocket launch($M)"].fillna(df["cost of the rocket launch($M)"].median(), inplace=True)
```

```python
df
```

```python
missing_values_after_cleaning = df.isnull().sum()
missing_values_after_cleaning
```

```python
print(df[df["Datum"].isna()])  # Show rows where Datum is NaT(Not A Time)
```

```python
#If missing values are few, you can drop them
df = df.dropna(subset=["Datum"])
```

```python
df.isnull().sum()
```

```python
# Check and remove duplicates
duplicates_count = df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Split 'Rocket name and satellites' into two columns: 'Rocket Name' and 'Satellite Name'
df[['Rocket Name', 'Satellite Name']] = df['Rocket name and satellites '].str.split('|', expand=True)
df['Rocket Name'] = df['Rocket Name'].str.strip()
df['Satellite Name'] = df['Satellite Name'].str.strip()
```

```python
# Drop original combined column
df = df.drop('Rocket name and satellites ', axis=1)
```

```python
# Split "Location" into "Launch Site & Space Center" and "Country"
df[["Launch Site & Space Center", "Country"]] = df["Location"].str.rsplit(",", n=1, expand=True)
```

```python
# Trim spaces from new columns
df["Launch Site & Space Center"] = df["Launch Site & Space Center"].str.strip()
df["Country"] = df["Country"].str.strip()
```

```python
# Display results
print(df[["Location", "Launch Site & Space Center", "Country"]].head())
```

```python
#Success Rate by Year (Trend Analysis)
df["Year"] = df["Datum"].dt.year  # Extract launch year

# Calculate yearly success rate
success_rate_by_year = df[df["Mission_Status"] == "Success"].groupby("Year").size() / df.groupby("Year").size()

# Plot success rate over time
plt.figure(figsize=(12,6))
success_rate_by_year.plot(marker="o", linestyle="-", color="green")
plt.title("Mission Success Rate Over the Years")
plt.xlabel("Year")
plt.ylabel("Success Rate")
plt.ylim(0,1)  # Scale from 0 to 1 (0% to 100%)
plt.grid()
plt.show()

```

## **Conclusions**

This project offered a comprehensive overview of global space launches by analyzing a detailed dataset containing information on rocket names, satellites, launch sites, costs, countries, companies, and mission outcomes. After extensive data cleaning and transformation, meaningful insights were extracted to better understand trends in the space industry.

1. **Industry Leadership**:  
   The majority of launches are conducted by a few leading countries, primarily the **USA**, **Russia**, and **China**. These nations have well-established government and private aerospace programs, reflecting their dominance in space operations.

2. **Emergence of Private Players**:  
   Companies like **SpaceX** have disrupted the traditional landscape by significantly increasing launch frequency and reducing costs, making space more accessible.

3. **Rocket Launch Costs Are Skewed**:  
   Most rocket launches cost under **100 million**,  but there are notable outliers with costs exceeding **$400 million**, often associated with large payloads or high-risk missions.

4. **High Mission Success Rates**:  
   Despite the complexity, the majority of missions are successful. This suggests robust technological and logistical execution, although failed missions still occur and warrant further investigation.

5. **Strategic Location of Launch Sites**:  
   Launch sites are often positioned near coastlines or equatorial regions, which helps maximize fuel efficiency and minimize risk, reinforcing the importance of location planning in aerospace strategy.

6. **Importance of Data Preparation**:  
   The success of this analysis was heavily dependent on proper data cleaning—removing duplicates, handling null values, correcting data types, and restructuring columns. Without this step, accurate insights would not have been possible.
