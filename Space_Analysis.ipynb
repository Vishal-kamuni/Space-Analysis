```python
# Drop irrelevant columns
df = df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
```

```python
# Rename Column
df = df.rename(columns={"Status Rocket": "Rocket_Status", " Rocket": "cost of the rocket launch($M)", "Status Mission": "Mission_Status","Detail":"Rocket name and satellites "})
```

```python
# Convert "Datum" to datetime format
df["Datum"] = pd.to_datetime(df["Datum"], errors='coerce')
```

```python
df.dtypes
```

```python
#Convert "Cost of the rocket launch" column to numeric
df["cost of the rocket launch($M)"] = pd.to_numeric(df["cost of the rocket launch($M)"], errors='coerce')
```

```python
#handle missing values
df["cost of the rocket launch($M)"].fillna(df["cost of the rocket launch($M)"].median(), inplace=True)
```

```python
df
```

```python
missing_values_after_cleaning = df.isnull().sum()
missing_values_after_cleaning
```

```python
print(df[df["Datum"].isna()])  # Show rows where Datum is NaT(Not A Time)
```

```python
#If missing values are few, you can drop them
df = df.dropna(subset=["Datum"])
```

```python
df.isnull().sum()
```

```python
# Check and remove duplicates
duplicates_count = df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Split 'Rocket name and satellites' into two columns: 'Rocket Name' and 'Satellite Name'
df[['Rocket Name', 'Satellite Name']] = df['Rocket name and satellites '].str.split('|', expand=True)
df['Rocket Name'] = df['Rocket Name'].str.strip()
df['Satellite Name'] = df['Satellite Name'].str.strip()
```

```python
# Drop original combined column
df = df.drop('Rocket name and satellites ', axis=1)
```

```python
# Split "Location" into "Launch Site & Space Center" and "Country"
df[["Launch Site & Space Center", "Country"]] = df["Location"].str.rsplit(",", n=1, expand=True)
```

```python
# Trim spaces from new columns
df["Launch Site & Space Center"] = df["Launch Site & Space Center"].str.strip()
df["Country"] = df["Country"].str.strip()
```

```python
# Display results
print(df[["Location", "Launch Site & Space Center", "Country"]].head())
```

```python
# Plot the top 10 companies by number of launches
plt.figure(figsize=(12,6))
sns.countplot(y=df["Company Name"], order=df["Company Name"].value_counts().index[:10], palette="viridis")
plt.title("Top 10 Companies by Number of Space Launches")
plt.xlabel("Number of Launches")
plt.ylabel("Company Name")
plt.show()
```

```python
#2.What is the average cost of a launch?
df["cost of the rocket launch($M)"].mean()
```

```python
#3.What is the success rate of missions
df["Mission_Status"].value_counts()
```

```python
#Success Rate by Year (Trend Analysis)
df["Year"] = df["Datum"].dt.year  # Extract launch year

# Calculate yearly success rate
success_rate_by_year = df[df["Mission_Status"] == "Success"].groupby("Year").size() / df.groupby("Year").size()

# Plot success rate over time
plt.figure(figsize=(12,6))
success_rate_by_year.plot(marker="o", linestyle="-", color="green")
plt.title("Mission Success Rate Over the Years")
plt.xlabel("Year")
plt.ylabel("Success Rate")
plt.ylim(0,1)  # Scale from 0 to 1 (0% to 100%)
plt.grid()
plt.show()

```

```python
#7.What is the trend of launch costs over time
df["Datum"].value_counts()
```

```python
# Group data by year and calculate average launch cost for each year
average_cost_by_year = df.groupby('Year')['cost of the rocket launch($M)'].mean()

# Create a line chart
plt.figure(figsize=(12, 6))  # Adjust figure size if needed
plt.plot(average_cost_by_year.index, average_cost_by_year.values, marker='o', linestyle='-')
plt.title('Trend of Launch Costs Over Time')
plt.xlabel('Year')
plt.ylabel('Average Launch Cost ($M)')
plt.grid(True)  # Add a grid for better readability
plt.show()
```

```python
#8What is the relationship between mission status and launch costs
mission_cost_relation = df.groupby('Mission_Status')['cost of the rocket launch($M)'].mean()
print(mission_cost_relation)
```
